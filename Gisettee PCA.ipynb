```python
import pandas as pd

# read data
gs_feature_df = pd.read_csv('gisette_train.data', header = None, delim_whitespace = True)
gs_feature_df
```

```python
gs_label = pd.read_csv('gisette_train.labels')
gs_label
```

```python
gs_feature_df = pd.concat([gs_label, gs_feature_df], axis = 1)
```

```python
# put data into own dataframe

feature_df = gs_feature_df.iloc[:, 1:]
label_df = gs_feature_df.iloc[:, 0]
```

```python
# use sklearn to standardise the data
# so that the data has a normal distribution

from sklearn.preprocessing import StandardScaler
import numpy as np

x = feature_df.values
x = StandardScaler().fit_transform(x)

np.mean(x),np.std(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_gs = pd.DataFrame(x, columns = feat_cols)
normalised_gs
```

```python
# choose the first 2 principal components

from sklearn.decomposition import PCA

pca_gs = PCA(n_components=2)
principalComponents_gs = pca_gs.fit_transform(x)

principal_gs_df = pd.DataFrame(data = principalComponents_gs
             , columns = ['principal component 1', 'principal component 2'])
principal_gs_df
```

```python
print('Explained variation per principal component: {}'.format(pca_gs.explained_variance_ratio_))
```

```python
# plot the data

import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component 1',fontsize=20)
plt.ylabel('Principal Component 2',fontsize=20)
plt.title("Principal Component Analysis of Gisettee DataSet",fontsize=20)
targets = [1, -1]
colors = ['b', 'r']
for target, color in zip(targets,colors):
    indicesToKeep = label_df == target
    plt.scatter(principal_gs_df.loc[indicesToKeep, 'principal component 1']
               , principal_gs_df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```

```python
# Shows how many principal components need to explain the variance in data

pca = PCA().fit(x)

evr = pca.explained_variance_ratio_
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(evr)))
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Total Explained Variance');
```
